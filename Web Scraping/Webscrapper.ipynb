## Extract relevant urls

```python
import re 

with open('NewsSources.txt', encoding="utf8") as f:
    urls = f.read()
    urls = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', urls)
i = 0
n = len(urls)
data = []
print(urls)
```

## Web Scrape for Articles

```python
import newspaper
for i in range(len(urls)):
    try:
        print("URL: " + urls[i])
        paper = newspaper.build(urls[i])
        for article in paper.articles:
            news = []
            article.download()
            article.parse()
            print(article.title)
            news.append(article.title)
            news.append(article.text)
            data.append(news)
    except:
        print("ERROR")
```
